```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
listing_data = pd.read_csv(r"C:\projects\paris_airbnb\listings.csv")
review_data = pd.read_csv(r"C:\projects\paris_airbnb\reviews.csv")
```

```python
complete_data = pd.merge(review_data, listing_data, left_on="listing_id", right_on="id", how="inner")
complete_data.head()
```

```python
complete_data.info()
```

```python
complete_data["date"] = pd.to_datetime(complete_data["date"])
complete_data["last_review"] = pd.to_datetime(complete_data["last_review"])
complete_data.dtypes
```

```python
complete_data.drop(columns=["neighbourhood_group"], inplace=True)
complete_data.columns
```

```python
no_of_duplicates = complete_data.duplicated().sum()
no_of_duplicates
```

```python
complete_data = complete_data.drop_duplicates()
complete_data.shape
```

```python
complete_data.select_dtypes(include="object").columns
```

```python
complete_data["host_name"].unique()
```

```python
complete_data["license"].isnull().sum()
```

```python
complete_data.drop(columns=["id", "host_name", "license", "last_review", "name"], inplace=True)
complete_data.info()
```

```python
complete_data.isnull().sum()
```

```python
median_value = complete_data["price"].median()
complete_data["price"].fillna(median_value, inplace=True)
```

```python
complete_data.isnull().sum()
```

```python
complete_data.head()
```

```python
complete_data.rename(columns={
                         'number_of_reviews': 'TotalReviews',
                         'reviews_per_month': 'MonthlyReviews',
                         'calculated_host_listings_count': 'TotalListingsByHost',
                         'availability_365': 'AvailabilityDays',
                         'number_of_reviews_ltm': 'ReviewsLast12Months' }, inplace=True)

complete_data.columns
```

```python
complete_data["price"] = complete_data["price"].round(0).astype(int)

complete_data["MonthlyReviews"] = complete_data["MonthlyReviews"].round(2)

complete_data["latitude"] = complete_data["latitude"].round(4)

complete_data["longitude"] = complete_data["longitude"].round(4)

complete_data.head()
```

```python
complete_data = complete_data.sort_values(by="date", ascending=False)
complete_data.head()
```

```python
complete_data.shape
```

```python
complete_data = complete_data.drop_duplicates(subset="listing_id", keep="first")
complete_data.head()
```

```python
complete_data.shape
```

```python
complete_data.to_csv(r"C:\projects\paris_airbnb\Paris Cleaned Dataset.csv",index=False)
```
